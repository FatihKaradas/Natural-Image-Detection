 Kütüphanelerin yüklenmesi

```python
import numpy as np
import matplotlib.pyplot as plt
import os, shutil
from keras import models
from keras import layers
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
from keras.applications import VGG16
```

Traning-Validation-Test klasörlerini oluşturma ve Verileri bu klasörlere gönderme işlemi


```python

base_dir = "/content/drive/MyDrive/data/natural_images"
class_names = ["airplane", "car", "cat", "dog", "flower", "fruit", "motorbike", "person"]

for split in ["Train", "Test", "Validation"]:
    split_dir = os.path.join(base_dir, split)
    os.makedirs(split_dir, exist_ok=True)

    for class_name in class_names:
        class_dir = os.path.join(split_dir, class_name)
        os.makedirs(class_dir, exist_ok=True)
image_counts = {
    "airplane": 727, "car": 968, "cat": 885, "dog": 702,
    "flower": 843, "fruit": 1000, "motorbike": 788, "person": 986
}

split_ratios = [0.6, 0.2, 0.2]

for class_name in class_names:
    class_data = f"/content/drive/MyDrive/data/natural_images/{class_name}"
    for i, split in enumerate(["Train", "Test", "Validation"]):
        start_idx = int(image_counts[class_name] * sum(split_ratios[:i]))
        end_idx = int(image_counts[class_name] * sum(split_ratios[:i+1]))
        for fname in os.listdir(class_data)[start_idx:end_idx]:
            src = os.path.join(class_data, fname)
            dst = os.path.join(base_dir, split, class_name, fname)
            shutil.copyfile(src, dst)
train_dir = "/content/drive/MyDrive/data/natural_images/Train"
test_dir = "/content/drive/MyDrive/data/natural_images/Test"
validation_dir = "/content/drive/MyDrive/data/natural_images/Validation"
```

Modelin minimum başarı oranının hesaplanması

```python
##common sense baseline
##((727/6899*968/6899*885/6899*702/6899*843/6899*1000/6899*788/6899*986/6899)+
##  (1-727/6899)*(1-968/6899)*(1-885/6899)*(1-702/6899)*(1-843/6899)*(1-1000/6899)*(1-788/6899)*(1-986/6899))*100=%40.70
print('Minimum başarı oranı: %40.70')
```

CNN modelin oluşturulması

```python
model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation = "relu", input_shape = (150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation = "relu"))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation = "relu"))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation = "relu"))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())

model.add(layers.Dense(512, activation = "relu"))
model.add(layers.Dense(256, activation = "relu"))
model.add(layers.Dense(128, activation = "relu"))
model.add(layers.Dense(64, activation = "relu"))
model.add(layers.Dense(32, activation = "relu"))
model.add(layers.Dense(16, activation = "relu"))
model.add(layers.Dense(8, activation = "softmax"))
```

Oluşturduğum CNN modelin özetinin çıktısı

```python
model.summary()
```

CNN modelin compile edilmesi

```python
model.compile(loss = "categorical_crossentropy", optimizer = optimizers.RMSprop(learning_rate = 1e-4), metrics = ["acc"])
```

Normalizasyon işlemi ve Networke hazır hale getirmek için batch shape yaptığım kısım

```python
train_datagen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size = (150, 150),
    batch_size = 20,
    class_mode = "categorical"
)

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size = (150, 150),
    batch_size = 20,
    class_mode = "categorical"
)
for data_batch, labels_batch in train_generator:
    print("data batch shape:", data_batch.shape)
    print("labels batch shape:", labels_batch.shape)
    break
```

Training başlatıldı

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch = 100,
    epochs = 50,
    validation_data = validation_generator,
    validation_steps = 50
)

model.save("natural_images_classification.h5")
```

Doğruluk değerimizi artırmak için modelimizi önceden oluşturulmuş VGG16 modeliyle eğiteceğiz.

```python
conv_base = VGG16(weights = "imagenet",
                  include_top = False,
                  input_shape = (150, 150, 3))

conv_base.summary()
```

```python
datagen = ImageDataGenerator(rescale = 1./255)
batch_size = 20

def extract_features(directory, sample_count):
    features = np.zeros(shape = (sample_count, 4, 4, 512))
    labels = np.zeros(shape = (sample_count, 8))

    generator = datagen.flow_from_directory(
        directory,
        target_size = (150, 150),
        batch_size = batch_size,
        class_mode = "categorical"
    )
    i = 0
    for inputs_batch, labels_batch in generator:
        features_batch = conv_base.predict(inputs_batch)
        features[i * batch_size : (i + 1) * batch_size] = features_batch
        labels[i * batch_size : (i + 1) * batch_size] = labels_batch
        i += 1
        if i * batch_size >= sample_count:
            break
    return features, labels
```

```python
train_features, train_labels = extract_features(train_dir, 4136)
test_features, test_labels = extract_features(test_dir, 1380)
validation_features, validation_labels = extract_features(validation_dir, 1383)
```

```python
train_features = np.reshape(train_features, (4136, 4 * 4 * 512))
test_features = np.reshape(test_features, (1380, 4 * 4 * 512))
validation_features = np.reshape(validation_features, (1383, 4 * 4 * 512))
```

```python
model = models.Sequential()
model.add(layers.Dense(512, activation = "relu", input_dim = 4 * 4 * 512))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(8, activation = "softmax"))

model.compile(loss = "categorical_crossentropy", optimizer = optimizers.RMSprop(learning_rate = 2e-5), metrics = ["acc"])
```

```python
history = model.fit(
    train_features, train_labels,
    epochs = 30,
    batch_size = 20,
    validation_data = (validation_features, validation_labels))

model.save("natural_images_classification_VGG16.h5")
```

```python
acc = history.history["acc"]
val_acc = history.history["val_acc"]
loss = history.history["loss"]
val_loss =  history.history["val_loss"]

epochs =  range(1, len(acc) + 1)

plt.plot(epochs, acc, "bo", label = "Training Accuracy")
plt.plot(epochs, val_acc, "b", label = "Validation Accuracy")
plt.title("Training and Validation Accuracy")
plt.legend()
plt.figure()
plt.plot(epochs, loss, "bo", label = "Training Loss")
plt.plot(epochs, val_loss, "b", label = "Validation Loss")
plt.title("Training and Validation Losses")
plt.legend()

plt.show()
```

```python
model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(8, activation='softmax'))

model.summary()
```

```python
conv_base.trainable = True

set_trainable = False
for layer in conv_base.layers:
    if layer.name == "block5_conv1":
        set_trainable = True
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False

model.compile(loss = "categorical_crossentropy", optimizer = optimizers.RMSprop(learning_rate = 1e-5), metrics = ["acc"])
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size = (150, 150),
    batch_size = 20,
    class_mode = "categorical"
)

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size = (150, 150),
    batch_size = 20,
    class_mode = "categorical"
)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size = (150, 150),
    batch_size = 20,
    class_mode = "categorical"
)
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch = 100,
    epochs = 50,
    validation_data = validation_generator,
    validation_steps = 50
)

model.save("natural_images_classification_Frozen.h5")
```

```python
acc = history.history["acc"]
val_acc = history.history["val_acc"]
loss = history.history["loss"]
val_loss =  history.history["val_loss"]

epochs =  range(1, len(acc) + 1)

plt.plot(epochs, acc, "bo", label = "Training Accuracy")
plt.plot(epochs, val_acc, "b", label = "Validation Accuracy")
plt.title("Training and Validation Accuracy")
plt.legend()
plt.figure()

plt.plot(epochs, loss, "bo", label = "Training Loss")
plt.plot(epochs, val_loss, "b", label = "Validation Loss")
plt.title("Training and Validation Losses")
plt.legend()

plt.show()
```
